#### Plotting the RMSE values for 100 particles, dt = 0.01 ps, total simulaton time 100 ps, using NVT Langevin as analysis and NVE Leapfrog as the reference. 

````python
import numpy as np
import matplotlib.pyplot as plt

#----------------------------------------------------------------------
# LOAD DATA AND SET PARAMETERS FOR PLOTTING
#----------------------------------------------------------------------
file1 = np.load("NVT_Langevin, dt = 0.01 ps_NVE_Leapfrog, dt = 0.01 ps,L=7nm_RMSE.npy")
file2 = np.load("NVT_Langevin, dt = 0.01 ps_NVE_Leapfrog, dt = 0.01 ps,L=15nm_RMSE.npy")
file3 = np.load("NVT_Langevin, dt = 0.01 ps_NVE_Leapfrog, dt = 0.01 ps,L=25nm_RMSE.npy")
file4 = np.load("NVT_Langevin, dt = 0.01 ps_NVE_Leapfrog, dt = 0.01 ps,L=35nm_RMSE.npy")
file5 = np.load("NVT_Langevin, dt = 0.01 ps_NVE_Leapfrog, dt = 0.01 ps,L=50nm_RMSE.npy")

dt_file1 = 0.01 # time-step of the simulation in the first file to plot [ps]
dt_file2 = 0.01 # time-step of the simulation in the second file to plot [ps]

total_time = 100 # total time in picoseconds
# DON'T FORGET TO CHANGE X,Y LABELS, TITLE AND NAME OF THE FILE!
#-----------------------------------------------------------------------
continue_ = str(input("Did you change all of the parameters and title correctly? (y/n) "))
if continue_ != "y":
    print("Check again the parameters!")
    raise SystemExit

n_steps_file1 = total_time/dt_file1 # number of steps
time_ps_file1 = np.arange(n_steps_file1 + 1) * dt_file1

n_steps_file2 = total_time/dt_file2 # number of steps
time_ps_file2 = np.arange(n_steps_file2 + 1) * dt_file2
#---------------------------------------------
# Defining plotting parameters
plt.rcParams.update({
    'font.size': 22,              # Standardgröße
    'axes.labelsize': 24,         # Achsenbeschriftung
    'xtick.labelsize': 22,        # X-Achse Werte
    'ytick.labelsize': 22,        # Y-Achse Werte
    'legend.fontsize': 22,        # Legendentext
})
#---------------------------------------------
plt.figure(figsize=(8, 6))
plt.plot(time_ps_file1, file1, label="L = 7 nm", color="lightsteelblue")
plt.plot(time_ps_file2, file2, label="L = 15 nm", color="cornflowerblue", ls = "-")
plt.plot(time_ps_file2, file3, label="L = 25 nm", color="dodgerblue", ls = "-")
plt.plot(time_ps_file2, file4, label="L = 35 nm", color="royalblue", ls = "-")
plt.plot(time_ps_file2, file5, label="L = 50 nm", color="navy", ls = "-")
plt.xlabel(r"$t\ \mathrm{[ps]}$")
plt.ylabel(r"$RMSE\ \mathrm{[nm]}$")
#plt.title("RMSE NVT_Langevin, reference NVE_Leapfrog, different L")
handles, labels = plt.gca().get_legend_handles_labels()
plt.legend(handles[::-1], labels[::-1], loc="upper left", bbox_to_anchor=(1.0, 1.035))
plt.tick_params(direction='in', which='both', top=True, bottom=True, left=True, right=True)
plt.savefig("RMSE NVT_Langevin, ref NVE_Leapfrog different L_final.png", dpi=300, bbox_inches='tight')
plt.show()
````

````python
#----------------------------------------------------------------
#   COMPARING ENERGY TRAJECTORIES ACROSS TIMESTEPS
#----------------------------------------------------------------
# The energy trajectory files have the following columns:
# E_pot  E_kin  T  P (index 0 to 3, respectively)

# Load energy trajectory files and parameters together as a list of tuples
simulations = [
    ("dt = 0.001 ps", "blue",   "sim_NVE_leapfrog_25nm_100particles_dt0p001_ene.npy",   "sim_NVE_leapfrog_25nm_100particles_dt0p001_params.npy"),
    ("dt = 0.002 ps", "orange", "sim_NVE_leapfrog_25nm_100particles_dt0.002_ene.npy",   "sim_NVE_leapfrog_25nm_100particles_dt0.002_params.npy"),
    ("dt = 0.003 ps", "green",  "sim_NVE_leapfrog_25nm_100particles_dt0.003_ene.npy",   "sim_NVE_leapfrog_25nm_100particles_dt0.003_params.npy"),
    ("dt = 0.004 ps", "red",    "sim_NVE_leapfrog_25nm_100particles_dt0p004_ene.npy",   "sim_NVE_leapfrog_25nm_100particles_dt0p004_params.npy"),
]

# Output settings
file_name_compare = "leapfrog_25nm_100particles_dt_comparison"
title_addition = " NVE Leapfrog, 25 nm, 100 particles"
save_figure = False

# Load all data into memory
data = []
for label, color, ene_file, params_file in simulations:
    ene = np.load(ene_file)
    params = np.load(params_file, allow_pickle=True).item()
    time_ps = np.arange(params["n_steps"] + 1) * params["dt"]
    data.append((label, color, ene, time_ps))

# Helper to reduce repetition in plotting
def make_comparison_plot(ylabel, title_suffix, col_indices, save_suffix):
    """
    col_indices: list of column indices to sum (e.g. [0] for E_pot, [0,1] for E_tot)
    """
    plt.figure(figsize=(10, 6))
    for label, color, ene, time_ps in data:
        y = sum(ene[:, i] for i in col_indices)
        plt.plot(time_ps, y, label=label, color=color)
    plt.xlabel("time [ps]")
    plt.ylabel(ylabel)
    plt.title(title_suffix + title_addition)
    plt.legend()
    if save_figure:
        plt.savefig(file_name_compare + save_suffix, dpi=300, bbox_inches='tight')
    plt.show()

# Potential energy
make_comparison_plot("$E_{pot}$ [kJ/mol]", "Potential Energy,", [0], "_Epot.png")

# Kinetic energy
make_comparison_plot("$E_{kin}$ [kJ/mol]", "Kinetic Energy,", [1], "_Ekin.png")

# Total energy
make_comparison_plot("$E_{tot}$ [kJ/mol]", "Total Energy,", [0, 1], "_Etot.png")

# Temperature
make_comparison_plot("T [K]", "Temperature,", [2], "_Temperature.png")

# Pressure
make_comparison_plot("P [Pa]", "Pressure,", [3], "_P.png")
````
